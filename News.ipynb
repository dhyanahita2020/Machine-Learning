```
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer  = TfidfVectorizer()
X = df.title
y = df.label
```

```
X = df.text
y =df.label
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size = 0.3)
```

```
tf_idf = TfidfVectorizer(stop_words = 'english').fit(X_train)
text_train = pd.DataFrame(tf_idf.transform(X_train).todense(), columns = tf_idf.get_feature_names())
text_test = pd.DataFrame(tf_idf.transform(X_test).todense(), columns = tf_idf.get_feature_names())
```

```
text_train
```

```
text_test
```

```
train = pd.concat([title_train, text_train], axis = 1)
test = pd.concat([title_test, text_test], axis =1)
```

```
print(train.shape)
print(test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix

model_bernb = BernoulliNB()

model_bernb.fit(train,y_train)
y_predict = model_bernb.predict(test)


accuracy_score(y_test,y_predict)
```
